```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 50
```

Partie 1

```python
# lire le CSV 
df = pd.read_csv("https://datasets.imdbws.com/title.akas.tsv.gz", sep = '\t')
```

```python
display(df)
print(df.shape)
```

```python
# compter le nombre de valeurs différentes de la colonne région  
count_region = df['region'].value_counts()
print(count_region)
```

```python
# pour ensuite filtrer uniquement la valeur FR 
condition = df['region'].str.contains('FR', case = False, na = False)
df_filtre =  df[condition]
display(df_filtre)
```

```python
# les ordonner par ordre croissant d'ID 
df_filtre = df_filtre.sort_values('titleId')
```

```python
display(df_filtre)
```

```python
# Test
print(df_filtre[df_filtre.titleId == "tt1675434"])
```

```python
#voir les modalité sur types
print(df_filtre.types.value_counts())
```

```python
#voir original title
print(df_filtre.isOriginalTitle.value_counts())
```

```python
#distribution des données dans language
print(df_filtre.language.value_counts())
```

```python
#filtrer les film sur language : fr et \\N
```

```python
df_semi_final = df_filtre[df_filtre.language.isin(["\\N", 'fr'])]
display(df_semi_final)
```

```python
#filtrer avec imdb display
condition1 = df_semi_final['types'].str.contains('imdbDisplay', case = False, na = False)
df_semi_final2 = df_semi_final[condition1]
```

```python
#drop cpolonnes region, ordering,attributes, isOriginalTitle, type
df_final = df_semi_final2.drop(columns = ["types","attributes", "isOriginalTitle", "ordering", "language", "region"]) 
display(df_final)
```
